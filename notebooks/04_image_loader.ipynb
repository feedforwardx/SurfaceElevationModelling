```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import os

img_path = "/kaggle/input/data/image.img"
output_png_path = "/kaggle/working/converted_image.png"

with rasterio.open(img_path) as src:
    img_data = src.read(1) 
    img_norm = (img_data - img_data.min()) / (img_data.max() - img_data.min())
    img_8bit = (img_norm * 255).astype(np.uint8)
    Image.fromarray(img_8bit).save(output_png_path)

plt.figure(figsize=(8, 8))
plt.imshow(img_8bit, cmap='gray')
plt.axis('off')
plt.title("Converted .img to PNG")
plt.show()
```
